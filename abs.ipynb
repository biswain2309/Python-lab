```python
working_template = load_workbook(filename='WORKING-tmp.xlsx')
sheetnames = working_template.get_sheet_names()
name = sheetnames[0]
sheet_ranges = working_template[name]
wt_df = pd.DataFrame(sheet_ranges.values)
wt_df.columns = wt_df.iloc[1]
wt_df = wt_df.drop([0,1])
# wt_df.iloc[0]['al_link']
wt_df
```

```python
number_of_rows_each_df = 1000000

df = pd.DataFrame(columns=['id', 'name','al_link', 'al_showslide2','al_pre-approvedText','al_loanAmount','twl_showslide2','twl_loanAmount','twl_link','pl_loanType', 'pl_pre-approvedText','pl_loanAmount','pl_link','hl_name','hl_link','instaloan_show','instaloan_link','instaloan_amount','instaloan_cardno','consumerLoan_showslide','consumerloan_amount','consumerloan_cardno','consumerloan_link','cc_cardno','cc_link','url'])
for i,j in input_data_1_2.iterrows():
#         print(type(j[i+5]))
#         print(j[i+5])
#         print(wt_df.iloc[0]['pl_preapprovedText'])
    df['id'] = 'a' + (random_excel.sample(1)['EBB9D209']).astype(str).str.lower()
    df['name'] = j[i].split(" ")[0]
    df['al_link'] = wt_df.iloc[0]['al_link'].replace("$D3",j[i+2])
    df['al_showslide2'] = wt_df.iloc[0]['al_showslide2'].replace("D3",j[i+2])
    df['al_pre-approvedText'] = wt_df.iloc[0]['al_pre-approvedText'].replace("$D3",j[i+2])
    df['al_loanAmount'] = j['AL_AMT']
    df['twl_showslide2'] = wt_df.iloc[0]['twl_showslide2'].replace("$J3",str(j[i+9]))
    df['twl_loanAmount'] = j['TWL_AMT']
    df['twl_link'] = 'https://leads.hdfcbank.com/applications/new_webforms/apply/mobile/Two-wheeler-Loan.aspx?pcode=QLS_VIDEO_VIS'
    df['pl_loanType'] = wt_df.iloc[0]['pl_loanType'].replace("$F3",str(j[i+5]))
    df['pl_pre-approvedText'] = wt_df.iloc[0]['pl_preapprovedText'].replace("$F3",str(j[i+5]))
    df['pl_loanAmount'] = j['PL_AMT']
    df['pl_link'] = wt_df.iloc[0]['pl_link'].replace("$F3",str(j[i+5]))
    df['hl_name'] = ""
    df['hl_link'] = 'https://leads.hdfcbank.com/applications/new_webforms/apply/mobile/Home-Loan-Indians.aspx?SourceCode=paid&pcode=QLS_EMAIL&promocode=QLS_EMAIL&utm_campaign=QLS&utm_medium=email&utm_source=QLS_EMAIL'
    df['instaloan_show'] = wt_df.iloc[0]['instaloan_show'].replace("$M3",str(j[i+12]))
    df['instaloan_link'] = wt_df.iloc[0]['instaloan_link']
    df['instaloan_amount'] = j[i+10]
    df['instaloan_cardno'] = 'XX' + str(j[i+12])
    df['consumerLoan_showslide'] = wt_df.iloc[0]['consumerLoan_showslide'].replace("$O3",str(j[i+14]))
    df['consumerloan_amount'] = j[i+14]
    df['consumerloan_cardno'] = j[i+15]
    df['consumerloan_link'] = 'https://v1.hdfcbank.com/htdocs/common/festive-treats/loan_product/easyemi.html?pcode=QLS_EMAIL&promocode=QLS_EMAIL&utm_campaign=%20QLS&utm_medium=email&utm_source=QLS_EMAIL'
    df['cc_cardno'] = ""
    df['cc_link'] = wt_df.iloc[0].replace("$R3",str(j[i+17]))
    df['url'] = wt_df.iloc[0].replace("$W3",str(j[i+22]))
    df.reset_index(drop=True, inplace=True)
    
    total_rows = df.index
   
    if df.index == number_of_rows_each_df:
        df.to_excel('output{}'.format()')

print(df)
```
